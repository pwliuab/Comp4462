```python
df = pd.read_csv('drive/MyDrive/athlete_events.csv')

# throw those null value data
removedNA_Age_Medal = df.dropna(subset=['Medal','Age'])

len(removedNA_Age_Medal)
```

```python
#separate Male and female into two dataframes
sex_grouped = df.groupby(removedNA_Age_Medal.Sex)
male = sex_grouped.get_group("M")
female = sex_grouped.get_group("F")
```

```python
year_grouped = df.groupby(removedNA_Age_Medal.Year)
year = year_grouped.get_group(1996)
yearSets = set()
year_hash_dic = {}

# use set to calculate the total number of years we got
# use list to store all the same year data
# use obj to store different year
# show all of the years 
for personal_id, personal_data in female.iterrows():
  yearSets.add(personal_data[9])
  year = personal_data[9]
  if year in year_hash_dic:
    continue
  else:
    print(year)
    year_hash_dic[year] = year_grouped.get_group(year)

```

```python
sorted_yearlist = list(sorted(yearSets))

for year in sorted_yearlist:
  print("Mean Age of year " + str(year) + " " + str(year_hash_dic[year]['Age'].mean()))
  print("Median Age of year " + str(year) + " " + str(year_hash_dic[year]['Age'].median()))
```

```python
for year in sorted_yearlist:
  sex_grouped = year_hash_dic[year].groupby('Sex')
  print("Male Mean Age of year " + str(year) + " " + str(sex_grouped.get_group('M')['Age'].mean()))
  print("Male Median Age of year " + str(year) + " " + str(sex_grouped.get_group('M')['Age'].median()))
  print("################################################################################################")
  print("Female Mean Age of year " + str(year) + " " + str(sex_grouped.get_group('F')['Age'].mean()))
  print("Female Median Age of year " + str(year) + " " + str(sex_grouped.get_group('F')['Age'].median()))
```

```python
# year female_mean_age male_mean_age female_median_age male_median_age overall_mean_age overall_median_age
result_dic = {}
firstTime = True
for year in sorted_yearlist:
  sex_grouped = year_hash_dic[year].groupby('Sex')
  if firstTime == True:
    result_dic['overall_mean_age'] = [year_hash_dic[year]['Age'].mean()]
    result_dic['overall_median_age'] = [year_hash_dic[year]['Age'].median()]
    result_dic['female_mean_age'] = [sex_grouped.get_group('F')['Age'].mean()]
    result_dic['female_median_age'] = [sex_grouped.get_group('F')['Age'].median()]
    result_dic['male_mean_age'] = [sex_grouped.get_group('M')['Age'].mean()]
    result_dic['male_median_age'] = [sex_grouped.get_group('M')['Age'].median()]
    result_dic['year'] = [year]
    firstTime = False
  else:
    result_dic['overall_mean_age'].append(year_hash_dic[year]['Age'].mean())
    result_dic['overall_median_age'].append(year_hash_dic[year]['Age'].median())
    result_dic['female_mean_age'].append(sex_grouped.get_group('F')['Age'].mean())
    result_dic['female_median_age'].append(sex_grouped.get_group('F')['Age'].median())
    result_dic['male_mean_age'].append(sex_grouped.get_group('M')['Age'].mean())
    result_dic['male_median_age'].append(sex_grouped.get_group('M')['Age'].median())
    result_dic['year'].append(year)
  print("Mean Age of year " + str(year) + " " + str(year_hash_dic[year]['Age'].mean()))
  print("Median Age of year " + str(year) + " " + str(year_hash_dic[year]['Age'].median()))
  print("Male Mean Age of year " + str(year) + " " + str(sex_grouped.get_group('M')['Age'].mean()))
  print("Male Median Age of year " + str(year) + " " + str(sex_grouped.get_group('M')['Age'].median()))
  print("################################################################################################")
  print("Female Mean Age of year " + str(year) + " " + str(sex_grouped.get_group('F')['Age'].mean()))
  print("Female Median Age of year " + str(year) + " " + str(sex_grouped.get_group('F')['Age'].median()))
```

```python
# year female_mean_age male_mean_age female_median_age male_median_age overall_mean_age overall_median_age
sex_dic = {}
firstTime = True
for year in sorted_yearlist:
  sex_grouped = year_hash_dic[year].groupby('Sex')
  if firstTime == True:
    sex_dic['sex'] = ['F'] 
    sex_dic['year'] = [year]
    sex_dic['mean_age'] = [sex_grouped.get_group('F')['Age'].mean()]
    sex_dic['median_age'] = [sex_grouped.get_group('F')['Age'].median()]
    ########################################################################
    sex_dic['sex'].append('M')
    sex_dic['median_age'].append(sex_grouped.get_group('M')['Age'].median())
    sex_dic['mean_age'].append(sex_grouped.get_group('M')['Age'].mean())
    sex_dic['year'].append(year)

    firstTime = False
  else:
    sex_dic['sex'].append('F')
    sex_dic['year'].append(year)
    sex_dic['mean_age'].append(sex_grouped.get_group('F')['Age'].mean())
    sex_dic['median_age'].append(sex_grouped.get_group('F')['Age'].median())
    ########################################################################
    sex_dic['sex'].append('M')
    sex_dic['mean_age'].append(sex_grouped.get_group('M')['Age'].mean())
    sex_dic['median_age'].append(sex_grouped.get_group('M')['Age'].median())
    sex_dic['year'].append(year)

print(sex_dic)
```

```python
# year female_mean_age male_mean_age female_median_age male_median_age overall_mean_age overall_median_age
medal_dic = {}
firstTime = True
for year in sorted_yearlist:
  medal_grouped = year_hash_dic[year].groupby('Medal')
  # Bronze, Gold, Silver
  if firstTime == True:
    medal_dic['year'] = [year]
    medal_dic['medal'] = ['Gold']
    medal_dic['mean_age'] = [medal_grouped.get_group('Gold')['Age'].mean()]
    medal_dic['median_age'] = [medal_grouped.get_group('Gold')['Age'].median()]
    ########################################################################
    medal_dic['year'].append(year)
    medal_dic['median_age'].append(medal_grouped.get_group('Silver')['Age'].median())
    medal_dic['mean_age'].append(medal_grouped.get_group('Silver')['Age'].mean())
    medal_dic['medal'].append('Silver')
    ########################################################################
    medal_dic['year'].append(year)
    medal_dic['median_age'].append(medal_grouped.get_group('Bronze')['Age'].median())
    medal_dic['mean_age'].append(medal_grouped.get_group('Bronze')['Age'].mean())
    medal_dic['medal'].append('Bronze')

    firstTime = False
  else:
    medal_dic['year'].append(year)
    medal_dic['median_age'].append(medal_grouped.get_group('Gold')['Age'].median())
    medal_dic['mean_age'].append(medal_grouped.get_group('Gold')['Age'].mean())
    medal_dic['medal'].append('Gold')
    ########################################################################
    medal_dic['year'].append(year)
    medal_dic['median_age'].append(medal_grouped.get_group('Silver')['Age'].median())
    medal_dic['mean_age'].append(medal_grouped.get_group('Silver')['Age'].mean())
    medal_dic['medal'].append('Silver')
    ########################################################################
    medal_dic['year'].append(year)
    medal_dic['median_age'].append(medal_grouped.get_group('Bronze')['Age'].median())
    medal_dic['mean_age'].append(medal_grouped.get_group('Bronze')['Age'].mean())
    medal_dic['medal'].append('Bronze')

print(medal_dic)
```

```python
medal_df = pd.DataFrame(data=medal_dic)
medal_df.to_csv('drive/MyDrive/' + 'seperateMedal.csv')
```
